```python

import numpy as np

import matplotlib.pyplot as plt

import copy
```

```python
p0 = np.random.uniform(0,1,1)
def pi0(p):
    return 1

k=100
x = np.random.binomial(1,p0,k)
y = np.cumsum(x)
print(p0)
print(x)
print(y)
```

```python
p = np.linspace(0,1,1000)
l = p**0*pi0(p)
l = l/np.sum(l)*1000
plt.plot(p,l)
plt.scatter(np.array([p0]),np.array([0]))
plt.show()
print('Prior mean', np.mean(l*p))

for i in range(k):
    p = np.linspace(0,1,1000)
    l = p**y[i]*(1-p)**(i+1-y[i])*pi0(p)
    l = l/np.sum(l)*1000
    plt.plot(p,l)
    plt.scatter(np.array([p0]),np.array([0]))
    plt.show()
    print(np.mean(l*p))
```
